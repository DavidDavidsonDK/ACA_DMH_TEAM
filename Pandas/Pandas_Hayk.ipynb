# # Logistic Regression using Sklearn

```python
df = df[(df.district == 'Arabkir') | (df.district == 'Center')]
df = df.drop(['street', 'building_type'], axis=1)
df.district[df.district == 'Center'] = 0
df.district[df.district == 'Arabkir'] = 1
df.district = df.district.astype(np.int)

df.condition[df.condition == 'newly repaired'] = 3
df.condition[df.condition == 'good'] = 2
df.condition[df.condition == 'zero condition'] = 1
df.condition = df.condition.astype(np.int)
df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn import metrics
from sklearn.cross_validation import cross_val_score
import patsy
```

```python
#df_train = df.iloc[0: int(np.floor(2 * len(df)/3))]
#df_test = df.iloc[int(np.floor(2 * len(df)/3)):]
```

```python
formula = 'district ~ price + condition + max_floor + num_rooms + area + num_bathrooms + floor + ceiling_height' 
# create a results dictionary to hold our regression results for easy analysis later        
results = {} 
```

```python
y,x = dmatrices(formula, data=df, return_type='dataframe')
print(y.columns)

model = sm.Logit(y,x)
res = model.fit()

results['Logit'] = [res, formula]
res.summary()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
model2 = LogisticRegression()
model2.fit(X_train, y_train)
predicted = model2.predict(X_test)
```

```python
probas = model2.predict_proba(X_test)
print (probas[:10])
```

```python
print (metrics.accuracy_score(y_test, predicted))
print (metrics.roc_auc_score(y_test, probas[:, 1]))
```

# Camparing with my Logistic Regression

```python
from logistic_regression import gradient_descent
from logistic_regression import logistic_predict
from logistic_regression import normalized_data
```

```python
def accuracy_score(Y_true, Y_predict):
    true_predicts = 0
    for i in range(len(Y_true)):
        if Y_true[i] == Y_predict[i]:
            true_predicts += 1
    return true_predicts / len(Y_true)
```

```python
X_logistic_train = np.matrix(X_train.values.tolist())
X_logistic_test = np.matrix(X_test.values.tolist())
Y_logistic_train = y_train.values.tolist()
Y_logistic_test = y_test.values.tolist()

#converting 0 to -1
for i in range(len(Y_logistic_train)):
    if Y_logistic_train[i] == [0.0]:
        Y_logistic_train[i] = -1.0
    else:
        Y_logistic_train[i] = 1.0
for i in range(len(Y_logistic_test)):
    if Y_logistic_test[i] == [0.0]:
        Y_logistic_test[i] = -1.0
    else:
        Y_logistic_test[i] = 1.0
Y_logistic_train = np.array(Y_logistic_train)
Y_logistic_test = np.array(Y_logistic_test)

#Normalize data
X = np.row_stack((X_logistic_train, X_logistic_test))
XNormalize = normalized_data(X)[0]
X_logistic_train = np.matrix(XNormalize[:X_logistic_train.shape[0]])
X_logistic_test = np.matrix(XNormalize[X_logistic_train.shape[0]:])

#fitting
beta = gradient_descent(X_logistic_train, Y_logistic_train, epsilon=1e-6, l=1, step_size=1e-2, max_steps=30)[0]
ypred_logistic = logistic_predict(beta, X_logistic_test)
log_accuracy = accuracy_score(Y_logistic_test, ypred_logistic)
print(log_accuracy)
```
